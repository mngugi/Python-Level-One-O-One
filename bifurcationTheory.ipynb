```python
import numpy as np
import matplotlib.pyplot as plt

# Define the model
def server_stability(x, lambda_value):
    return lambda_value - x**2

# Real data for traffic rates (lambda) over time
time = np.array([1, 2, 3, 4, 5])  # Time (in seconds)
lambda_values = np.array([10, 20, 30, 50, 80])  # Traffic rates

# Simulate the system
x = 0  # Initial stability
x_values = []  # To store stability values

for i, lambda_value in enumerate(lambda_values):
    dx_dt = server_stability(x, lambda_value)
    x += dx_dt * 1  # Euler method (approximation for small time steps)
    x_values.append(x)

# Plot the results
plt.plot(time, x_values, label="Server Stability (x)")
plt.plot(time, lambda_values, label="Traffic Rate (λ)", linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Values")
plt.legend()
plt.title("Server Stability vs Traffic Rate")
plt.show()
```

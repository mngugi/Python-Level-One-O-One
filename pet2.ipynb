```python
from random import randrange

class Pet():
    boredom_decrement = 4
    hunger_decrement = 6
    boredom_threshold = 5
    hunger_threshold = 10
    sounds = ['Mrrp']
    
    def __init__(self, name="Kitty"):
        self.name = name
        self.hunger = randrange(self.hunger_threshold)
        self.boredom = randrange(self.boredom_threshold)
        self.sounds = self.sounds[:]
        
    def clock_tick(self):
        self.boredom += 1
        self.hunger += 1
        
    def mood(self):
        if self.hunger <= self.hunger_threshold and self.boredom <= self.boredom_threshold:
            return "happy"
        elif self.hunger > self.hunger_threshold:
            return "hungry"
        else:
            return "bored"
        
    def __str__(self):
        state = "I'm " + self.name + ", "
        state += "I feel " + self.mood() + ", "
        return state
    
    def hi(self):
        print(self.sounds[randrange(len(self.sounds))])
        self.reduce_boredom()
        
    def teach(self, word):
        self.sounds.append(word)
        self.reduce_boredom()
        
    def feed(self):
        self.reduce_hunger()
        
    def reduce_hunger(self):
        self.hunger = max(0, self.hunger - self.hunger_decrement)
        
    def reduce_boredom(self):
        self.boredom = max(0, self.boredom - self.boredom_decrement)
        
# anew class called Cat that inherits from class Pet

class Cat(Pet):
    sounds = ['Meow']

    def chasing_rats(self):
        return "What are you doning you pinky"
     

class Cheshire(Cat):
    def smile(self):
        print(":D :D :D")     

p1 = Pet("dougi")
print(p1)

for i in range(10):
    p1.clock_tick()
    print(p1)
p1.feed()
p1.hi()
p1.teach("Boo")

for i in range(10):
    p1.hi()
print(p1)        


Cat1 = Cat("PepperPinky")
print(Cat1)

Cat1.feed()
Cat1.hi()
print(Cat1)
print(Cat1.chasing_rats())

new_cat = Cheshire("Murshing")
new_cat.hi()
new_cat.chasing_rats()
new_cat.smile()

p1 = Pet("Lappers")
p1.hi()
```

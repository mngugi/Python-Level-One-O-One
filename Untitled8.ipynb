```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
import seaborn as sns
from scipy import stats
import statsmodels.formula.api as smf

# Define the file path


# Define the file path
path = '/home/x38fed/Documents/GitHub/DataScience/AvgPulseCalorieBurnage.csv'

# Read the CSV file into a DataFrame
dt = pd.read_csv(path)

# Set pandas options to display the entire DataFrame in a single cell
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Convert object columns to float64
dt["Average_Pulse"] = dt["Average_Pulse"].astype(float)
dt["Max_Pulse"] = dt["Max_Pulse"].astype(float)

# Print the first 10 rows of the DataFrame as a table
print("First 10 rows of the DataFrame:")
print(tabulate(dt.head(10), headers='keys', tablefmt='pretty'))

# Print summary statistics as a table
print("Summary Statistics:")
summary_stats = dt.describe()
print(tabulate(summary_stats, headers='keys', tablefmt='pretty'))

# Print DataFrame info
print("DataFrame Info:")
info = dt.info()
print(info)

# Plot Average_Pulse against Calorie_Burnage using matplotlib
dt.plot(x='Average_Pulse', y='Calorie_Burnage', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.show()

# Define a function to calculate slope
def slope(x1, y1, x2, y2):
    slope = (y2 - y1) / (x2 - x1)
    return slope

# Call the function with individual values
x1, y1, x2, y2 = 80, 240, 90, 260
result = slope(x1, y1, x2, y2)
print("-------+-----------")
print("Slope is:", result)
print("-------+-----------")

# Calculate slope and intercept using numpy.polyfit
slopIntercept = np.polyfit(dt["Average_Pulse"], dt["Calorie_Burnage"], 1)
print("-------+---+--------------------")
print("Slope at Intercept:", slopIntercept)
print("-------+------------------------")

# Define a function to predict calorie burnage
def function(x):
    return 2 * x + 80

print("-------+-------------------------------")
print("Calorie Burnage at 135 is:", function(135))
print("--------+-------------------------------")

# Calculate percentiles
print("-------+--------")
Max_Pulse = dt["Max_Pulse"]
twentyFivePercent = np.percentile(Max_Pulse, 25)
print("25% is:", twentyFivePercent)
fiftyPercent = np.percentile(Max_Pulse, 50)
print("50% is:", fiftyPercent)
seventyFivePercent = np.percentile(Max_Pulse, 75)
print("75% is:", seventyFivePercent)
tenPercent = np.percentile(Max_Pulse, 10)
print("10% is:", tenPercent)
print("-------+--------")

# Calculate standard deviation
print("-------+-----------------------------")
dtStandardDev = np.std(dt)
print("Standard Deviation:\n", dtStandardDev)
print("--------+----------------------------")

# Calculate mean
print("--------+--------")
mean = np.mean(dt)
print("mean", mean)
print("--------+--------")

# Calculate coefficient of variation
print("--------+----------------------------")
coefficientVar = dtStandardDev / mean
print("Coefficient Variable:\n", coefficientVar)
print("---------+---------------------------")

# Calculate variance
print("--------+--------")
variance = np.var(dt)
print("Variance is: \n", variance)
print("\n")
print("--------+--------")

# Correlation matrix
CorrelationMatrix = dt.corr()
print("Correlation Matrix is:\n", CorrelationMatrix)

# Plot correlation heatmap
axis_corr = sns.heatmap(CorrelationMatrix, vmin=-0.5, vmax=0.5, center=0,
                       cmap=sns.diverging_palette(30, 100, n=600),
                       square=True)
plt.title("Correlation Heatmap")
plt.show()

# Calculate correlation between Calorie_Burnage and Average_Pulse
correlation = dt["Calorie_Burnage"].corr(dt["Average_Pulse"])

# Plot scatter plot
dt.plot(x="Calorie_Burnage", y="Average_Pulse", kind="scatter")
plt.title(f"Scatter Plot (Correlation: {correlation:.2f})")
plt.xlabel("Calorie_Burnage")
plt.ylabel("Average_Pulse")
plt.grid(True)
plt.show()

# Linear regression
x = dt["Average_Pulse"]
y = dt["Calorie_Burnage"]
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")
plt.show()



# Read the CSV file into a DataFrame
dt = pd.read_csv(path)

# Set pandas options to display the entire DataFrame in a single cell
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Convert object columns to float64
dt["Average_Pulse"] = dt["Average_Pulse"].astype(float)
dt["Max_Pulse"] = dt["Max_Pulse"].astype(float)

# Print the first 10 rows of the DataFrame as a table
print("First 10 rows of the DataFrame:")
print(tabulate(dt.head(10), headers='keys', tablefmt='pretty'))

# Print summary statistics as a table
print("Summary Statistics:")
summary_stats = dt.describe()
print(tabulate(summary_stats, headers='keys', tablefmt='pretty'))

# Print DataFrame info
print("DataFrame Info:")
info = dt.info()
print(info)

# Plot Average_Pulse against Calorie_Burnage using matplotlib
dt.plot(x='Average_Pulse', y='Calorie_Burnage', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.show()

# Define a function to calculate slope
def slope(x1, y1, x2, y2):
    slope = (y2 - y1) / (x2 - x1)
    return slope

# Call the function with individual values
x1, y1, x2, y2 = 80, 240, 90, 260
result = slope(x1, y1, x2, y2)
print("-------+-----------")
print("Slope is:", result)
print("-------+-----------")

# Calculate slope and intercept using numpy.polyfit
slopIntercept = np.polyfit(dt["Average_Pulse"], dt["Calorie_Burnage"], 1)
print("-------+---+--------------------")
print("Slope at Intercept:", slopIntercept)
print("-------+------------------------")

# Define a function to predict calorie burnage
def function(x):
    return 2 * x + 80

print("-------+-------------------------------")
print("Calorie Burnage at 135 is:", function(135))
print("--------+-------------------------------")

# Calculate percentiles
print("-------+--------")
Max_Pulse = dt["Max_Pulse"]
twentyFivePercent = np.percentile(Max_Pulse, 25)
print("25% is:", twentyFivePercent)
fiftyPercent = np.percentile(Max_Pulse, 50)
print("50% is:", fiftyPercent)
seventyFivePercent = np.percentile(Max_Pulse, 75)
print("75% is:", seventyFivePercent)
tenPercent = np.percentile(Max_Pulse, 10)
print("10% is:", tenPercent)
print("-------+--------")

# Calculate standard deviation
print("-------+-----------------------------")
dtStandardDev = np.std(dt)
print("Standard Deviation:\n", dtStandardDev)
print("--------+----------------------------")

# Calculate mean
print("--------+--------")
mean = np.mean(dt)
print("mean", mean)
print("--------+--------")

# Calculate coefficient of variation
print("--------+----------------------------")
coefficientVar = dtStandardDev / mean
print("Coefficient Variable:\n", coefficientVar)
print("---------+---------------------------")

# Calculate variance
print("--------+--------")
variance = np.var(dt)
print("Variance is: \n", variance)
print("\n")
print("--------+--------")

# Correlation matrix
CorrelationMatrix = dt.corr()
print("Correlation Matrix is:\n", CorrelationMatrix)

# Plot correlation heatmap
axis_corr = sns.heatmap(CorrelationMatrix, vmin=-0.5, vmax=0.5, center=0,
                       cmap=sns.diverging_palette(30, 100, n=600),
                       square=True)
plt.title("Correlation Heatmap")
plt.show()

# Calculate correlation between Calorie_Burnage and Average_Pulse
correlation = dt["Calorie_Burnage"].corr(dt["Average_Pulse"])

# Plot scatter plot
dt.plot(x="Calorie_Burnage", y="Average_Pulse", kind="scatter")
plt.title(f"Scatter Plot (Correlation: {correlation:.2f})")
plt.xlabel("Calorie_Burnage")
plt.ylabel("Average_Pulse")
plt.grid(True)
plt.show()

# Linear regression
x = dt["Average_Pulse"]
y = dt["Calorie_Burnage"]
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")
plt.show()

```

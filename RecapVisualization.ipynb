```python
# bargraph
import altair as alt
import pandas as pd

source = pd.DataFrame({
    'a': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'b': [28, 55, 43, 91, 81, 53, 19, 87, 52]
})

alt.Chart(source).mark_bar().encode(
    x='a',
    y='b'
)

```

```python
# linegraph
import altair as alt
import pandas as pd


source2 = pd.DataFrame ({ 
    'j': ['Lions', 'Elephants', 'Giraffs', 'Rhino', 'Buffalo'],
    'i': [2000,10000,5600, 1500, 11000]
    
})

alt.Chart(source2).mark_line().encode( x = 'j', y = 'i')
```

```python
import altair as alt
from vega_datasets import data

source = data.movies.url

alt.Chart(source).mark_bar().encode(
    alt.X("IMDB_Rating:Q", bin=True),
    y='count()',
)
```

```python
import altair as alt
from vega_datasets import data

# Data generators for the background
sphere = alt.sphere()
graticule = alt.graticule()

# Source of land data
source = alt.topo_feature(data.world_110m.url, 'countries')

# Layering and configuring the components
alt.layer(
    alt.Chart(sphere).mark_geoshape(fill='skyblue'),
    alt.Chart(graticule).mark_geoshape(stroke='ivory', strokeWidth=0.5),
    alt.Chart(source).mark_geoshape(fill='green', stroke='black')
).project(
    'naturalEarth1'
).properties(width=600, height=400).configure_view(stroke=None)
```

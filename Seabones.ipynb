```python
"""
Script for creating a Seaborn relational plot with the "tips" dataset.
"""

import seaborn as sns

# Set the Seaborn style to "whitegrid" for your plots
sns.set_style("whitegrid")

# Load the "tips" dataset provided by Seaborn
tips = sns.load_dataset("tips")

# Create a relational plot (scatter plot) using Seaborn's relplot function
sns.relplot(
    data=tips,          # The dataset to use
    x="total_bill",     # The x-axis data
    y="tip",            # The y-axis data
    col="time",         # Separate plots into columns based on the "time" column
    hue="smoker",       # Color the points based on the "smoker" column
    style="smoker",     # Differentiate the points with different markers based on "smoker"
    size="size"         # Vary the point sizes based on the "size" column
)
```
